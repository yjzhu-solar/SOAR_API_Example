```python
def eui_fsi_query(start_time, end_time,):
    eui_query = "SELECT data.begin_time, data.descriptor, data.data_item_oid, eui.cr_val1, eui.cr_val2, eui.c_delt1, eui.c_delt2, eui.rsun_arc, eui.num_pixels_per_axis, eui.dimension_index " + \
    "FROM soar.v_sc_data_item data INNER JOIN soar.v_eui_sc_fits eui ON data.data_item_oid = eui.data_item_oid " + \
    f"WHERE data.instrument = 'EUI' AND data.level = 'L2' AND data.begin_time>='{start_time}' AND data.begin_time<='{end_time}' " + \
    "AND (data.descriptor='eui-fsi174-image' OR data.descriptor='eui-fsi304-image') " + \
    "ORDER BY data.begin_time ASC"

    results = soar.launch_job_async(eui_query)
    results = results.get_results().to_pandas()

    results1 = results[results["dimension_index"] == 1]
    results1 = results1[['begin_time', 'descriptor', 'data_item_oid', 'cr_val1', 'c_delt1', 'rsun_arc', 'num_pixels_per_axis']]
    results1["pixel_size1"] = results1["c_delt1"]/results1["rsun_arc"]*const.R_sun.to_value(u.km)
    results1["fov1"] = results1["num_pixels_per_axis"]*results1["pixel_size1"]/1e3
    results1.drop("num_pixels_per_axis", axis=1, inplace=True)

    results2 = results[results["dimension_index"] == 2]
    results2 = results2[['data_item_oid', 'cr_val2', 'c_delt2', 'rsun_arc', 'num_pixels_per_axis']]
    results2["pixel_size2"] = results2["c_delt2"]/results2["rsun_arc"]*const.R_sun.to_value(u.km)
    results2["fov2"] = results2["num_pixels_per_axis"]*results2["pixel_size2"]/1e3
    results2.drop(["num_pixels_per_axis", "rsun_arc"], axis=1, inplace=True)

    results_final = pd.merge(results1, results2, on="data_item_oid", how="inner")
    return results_final
```
